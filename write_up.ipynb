```python
import numpy as np
from matplotlib import pyplot as plt

from randomwalker import coord
```

## Part 1

```python
step_size=3
coord(step_size)
```

```python
n_timesteps = 100
x = np.empty(n_timesteps, float)
y = np.empty(n_timesteps, float)
#a = np.zeros(n_timesteps, float)

for i in range(n_timesteps):
    
    if i == 0:
        x[0] = 0
        y[0] = 0
        plt.plot(x[i], y[i], 'r-')

    elif i > 0:
        plt.plot(x[i], y[i], 'ro')
        x[i] = coord(step_size) + x[i-1]
        y[i] = coord(step_size) + y[i-1]

#print(x, np.max(x), np.min(x), np.max(y), np.min(y))
bigx = np.absolute(x)
bigy = np.absolute(y)

xmax = np.max(bigx)
ymax = np.max(bigy)
print(xmax, ymax)
plt.xlim(-xmax, xmax)
plt.ylim(-ymax, ymax)       

#plt.plot(x[-1], y[-1], 'ro')   
        
plt.show()  

print(x[n_timesteps-1], y[n_timesteps - 1])
```
